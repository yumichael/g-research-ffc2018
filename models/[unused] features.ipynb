```python
def _inf_market(dh):
    return pd.get_dummies(dh[['Market']], columns=['Market'])
def _inf_stock(dh):
    return pd.get_dummies(dh[['Stock']], columns=['Stock'])
def _inf_day(dh):
    return pd.get_dummies(dh[['Day']], columns=['Day'])
```

```python
def _inf_dayseg(dh):
    dayseg = dh['Day']
    off0 = (dayseg + 2) // 14
    off0.name = 'Biweek0'
    off1 = (dayseg + 2 + 7) // 14
    off1.name = 'Biweek1'
    dayseg = pd.concat([off0, off1], axis=1)
    return dayseg
```

```python
def _inf_weekday(dh):
    wkdfeats = 'Monday Tuesday Wednesday Thursday Friday'.split()
    weekday = ((dh.Day + 2) % 7)
    weekday = weekday.map({i: s for i, s in enumerate(wkdfeats)})
    weekday = pd.get_dummies(weekday)
    weekday = weekday[wkdfeats]
    return weekday
```

```python
def _inf_count(dh):
    count = dh.groupby(['Market', 'Stock']).Stock.transform('count')
    count.name = '$ count'
```
