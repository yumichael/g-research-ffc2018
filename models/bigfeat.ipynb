#### window

```python
for F, C in itertools.product(['x3', 'xx'], [-93]):
    gc.collect()
    print(F, C, 'BEGIN')
    save_dir = big_dir + 'data/features/' # <1////////////
    feats = O(xx=xxfeats, x3=x3feats)[F] # <2=======================================
    names = combos.name.index
    dhlin = dh[feats]
    dhlog = np.log(dhlin.replace(0, np.exp(C))) # <5=======================
    thedh = dhlin # <4=========================================
    fname = F# + '.log.zc(' + str(C) + ')' # <24=========================================
    pathname = save_dir + '{}' + fname + '{}'
    ######################################################################
    grp = dh.Stock
    grpby = thedh.groupby(grp)
    tots = []
    _sym_ = '{}'
    print('# flat and before')
    for n in [1, 2, 3, 4, 5, 7, 9]:
        print(n)
        form = f'.window(<{n}>!{_sym_})'
        avg = grpby.apply(lambda x: x.rolling(n, 1).mean().shift(1))
        cnt = grpby.apply(lambda x: x.rolling(n, 1).count().shift(1))
        plus = (avg * cnt + thedh) / (cnt + 1)
        diff = thedh - avg
        ratio = thedh / avg
        guys = [df for df in [plus, diff, ratio]]
        for df, sym in zip(guys, '+ - /'.split()):
            df.columns = [s + form.format(sym) for s in df.columns]
        tots += guys
    print('# flat and after')
    for n in [1, 2, 3, 4, 5, 7, 9]:
        print(n)
        form = f'.window(!<{n}>{_sym_})'
        avg = grpby.apply(lambda x: x.rolling(n, 1).mean().shift(-n))
        cnt = grpby.apply(lambda x: x.rolling(n, 1).count().shift(-n))
        plus = (avg * cnt + thedh) / (cnt + 1)
        diff = thedh - avg
        ratio = thedh / avg
        guys = [df for df in [plus, diff, ratio]]
        for df, sym in zip(guys, '+ - /'.split()):
            df.columns = [s + form.format(sym) for s in df.columns]
        tots += guys
    print('# flat and middle')
    for n in [3, 5, 7, 9, 11]:
        print(n)
        form = f'.window(<<{n}>>{_sym_})'
        avg = grpby.apply(lambda x: x.rolling(n, 3, center=True).mean())
        cnt = grpby.apply(lambda x: x.rolling(n, 3, center=True).count())
        plus = avg + thedh * 0
        diff = (thedh - avg) * cnt / (cnt - 1)
        ratio = thedh / (thedh - diff)
        std = grpby.apply(lambda x: x.rolling(n, 3, center=True).std())
        stdr = thedh / std
        guys = [df for df in [plus, diff, ratio, std, stdr]]
        for df, sym in zip(guys, '+ - / ~ ~/'.split()):
            df.columns = [s + form.format(sym) for s in df.columns]
        tots += guys
    print('# triangle and middle')
    for n in [3, 5, 7, 9, 11, 19]:
        print(n)
        form = f'.window(<^{n}^>{_sym_})'
        avg = grpby.apply(lambda x: x.rolling(n, 3, center=True, win_type='triang').mean())
        sum = grpby.apply(lambda x: x.rolling(n, 3, center=True, win_type='triang').sum())
        plus = avg + thedh * 0
        other = (sum - thedh) / (sum / avg - 1)
        diff = thedh - other
        ratio = thedh / other
        guys = [df for df in [plus, diff, ratio]]
        for df, sym in zip(guys, '+ - /'.split()):
            df.columns = [s + form.format(sym) for s in df.columns]
        tots += guys
    print('# exp and before')
    for n in [2, 3, 4, 5, 7, 10, 15]:
        print(n)
        form = f'.window(<{n}>*!{_sym_})'
        alpha = 2 / (n + 1)
        avg = grpby.apply(lambda x: x.ewm(span=n, min_periods=1, adjust=False).mean().shift(1))
        plus = thedh * alpha + avg * (1 - alpha)
        diff = thedh - avg
        ratio = thedh / avg
        guys = [df for df in [plus, diff, ratio]]
        for df, sym in zip(guys, '+ - /'.split()):
            df.columns = [s + form.format(sym) for s in df.columns]
        tots += guys
    #####################################################################
    the = pd.concat(tots, axis=1).astype(np.float32)
    del tots; gc.collect()
    #thing.to_hdf(save_dir + '64/' + fname + '.window.hdf5', key=fname + '.window', mode='w')
    the.to_hdf(pathname.format('window/il32/', '.window'), key='a', mode='w')
    the.makesw().astype(np.float32).to_hdf(pathname.format('window/sw32/', '.window'), key='a', mode='w')
    ###
    o = O(); def_W(o); def_X(o, the, include='wqtl')
    del the; gc.collect()
    the = o.Xwqtl.astype(np.float32)
    del o; gc.collect()
    the.to_hdf(pathname.format('window/sw32/', '.window.wqtl'), key='a', mode='w')
    the.makeil_like(dh).to_hdf(pathname.format('window/il32/', '.window.wqtl'), key='a', mode='w')
    del the; gc.collect()
    print(F, C, 'END')
```

```python
for xname, theX in zip(['xo.log.z(1)', 'x3.log.zc(-93).mintco0[2,1]', 'xg.log.zc(-93).mintco0[2,1]'],
                       [o.Xlogz1, X3int, Xgint]):
    gc.collect()
    print(xname, 'BEGIN')
    save_dir = big_dir + 'data/features/' # <1////////////
    pathname = save_dir + '{}' + xname + '{}'
    ######################################################################
    tots = []
    print('# exp and before')
    for n in [5, 20]:
        print(n)
        form = '.window(<{}>*!{})'
        alpha = 2 / (n + 1)
        avg = theX.ewm(span=n, min_periods=1, adjust=False).mean().shift(1)
        plus = 0#(theX * alpha + avg * (1 - alpha)).colname_append(form.format(n, '+'), level=0)
        diff = (theX - avg).colname_append(form.format(n, '-'), level=0)
        tots += [diff]
        del avg, plus, diff; gc.collect()
    print('# exp and after')
    for n in [5, 20]:
        print(n)
        form = '.window(!*<{}>{})'
        alpha = 2 / (n + 1)
        theX.sort_index(ascending=False, inplace=True)
        avg = theX.ewm(span=n, min_periods=1, adjust=False).mean().shift(1)
        theX.sort_index(ascending=True, inplace=True)
        avg.sort_index(ascending=True, inplace=True)
        plus = 0#(theX * alpha + avg * (1 - alpha)).colname_append(form.format(n, '+'), level=0)
        diff = (theX - avg).colname_append(form.format(n, '-'), level=0)
        tots += [diff]
        del avg, plus, diff; gc.collect()
    #####################################################################
    the = pd.concat(tots, axis=1).astype(np.float32)
    del tots; gc.collect()
    #the.to_hdf(pathname.format('window/sw32/', '.winexp'), key='a', mode='w')
    ###
    meh = O(); def_W(meh); def_X(meh, the.makeil_like(dh), include='wqtl wnrm')
    the.astype(np.float32).to_hdf(pathname.format('window/sw32/', '.windiff'), key='a', mode='w')
    del the; gc.collect()
    meh.Xwqtl.astype(np.float32).to_hdf(pathname.format('window/sw32/', '.windiff.wqtl'), key='a', mode='w')
    meh.Xwnrm.astype(np.float32).to_hdf(pathname.format('window/sw32/', '.windiff.wnrm'), key='a', mode='w')
    del meh; gc.collect()
    print(xname, 'END')
```

```python
for xname, theX in zip(['xo.log.z(1)', 'x3.log.zc(-93).mintco0[2,1]', 'xg.log.zc(-93).mintco0[2,1]'],
                       [o.Xlogz1, X3int, Xgint]):
    gc.collect()
    print(xname, 'BEGIN')
    save_dir = big_dir + 'data/features/' # <1////////////
    pathname = save_dir + '{}' + xname + '{}'
    ######################################################################
    tots = []
    print('# gaussian')
    for n in [5, 20]:
        print(n)
        form = '.window(<@{}@>{})'
        avg = theX.rolling(3 * n, min_periods=3, center=True, win_type='gaussian').mean(std=n / 2)
        sum = theX.rolling(3 * n, min_periods=3, center=True, win_type='gaussian').sum(std=n / 2)
        plus = (avg + theX * 0).colname_append(form.format(n, '+'), level=0)
        other = (sum - theX) / (sum / avg - 1)
        diff = (theX - other).colname_append(form.format(n, '-'), level=0)
        tots += [plus]
    #####################################################################
    the = pd.concat(tots, axis=1).astype(np.float32)
    del tots; gc.collect()
    ###
    the.to_hdf(pathname.format('window/sw32/', '.nrmplus'), key='a', mode='w')
    the = the.makeil_like(dh)
    the.to_hdf(pathname.format('window/il32/', '.nrmplus'), key='a', mode='w')
    meh = O(); def_W(meh); def_X(meh, the, include='wqtl')
    del the; gc.collect()
    the = meh.Xwqtl.astype(np.float32)
    the.to_hdf(pathname.format('window/sw32/', '.nrmplus.wqtl'), key='a', mode='w')
    the.makeil_like(dh).to_hdf(pathname.format('window/il32/', '.nrmplus.wqtl'), key='a', mode='w')
    del the, meh; gc.collect()
    print(xname, 'END')
```
